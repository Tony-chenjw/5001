```python
import pandas as pd
from collections import Counter
pd.set_option('display.max_columns', None)
```

```python
total = pd.read_csv('./train_features.csv')
train_x = total.drop('speed', axis=1)
speed = total['speed']
train_x
```

```python
test_x = pd.read_csv('./test_features.csv')
test_x
```

```python
total = pd.concat([train_x, test_x], axis=0)
total
```

```python
total['year'] = total['year'].replace({2017:0, 2018:1})
total
```

```python
train_x = total[:14006]
train_x = pd.concat([train_x, speed], axis=1)
train_x
```

```python
test_x = total[14006:]
test_x
```

```python
train_x.to_csv('./train_features.csv', encoding='utf-8', index=False)
test_x.to_csv('./test_features.csv', encoding='utf-8', index=False)
```
